```python
categorical_columns = bmw_sales_df.select_dtypes(include='object').columns
numerical_columns = bmw_sales_df.select_dtypes(exclude='object').columns
```

```python
for col in categorical_columns:
    counts = bmw_sales_df[col].value_counts()
    plt.figure(figsize=(7, 4))
    if len(counts) <= 5:
        plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
        plt.title(f'{col} (pie chart)')
    else:
        sns.countplot(x=col, data=bmw_sales_df, order=counts.index)
        plt.title(f'{col} (count plot)')
        plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
